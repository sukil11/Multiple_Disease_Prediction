```python
df['Dataset'] = df['Dataset'].replace([2,1],[1,0])
df['Dataset'].head()
```

```python
df = pd.get_dummies(df, columns = ['Gender'], drop_first = True)
df.head()
```

```python
x = df.drop('Dataset', axis = 1)
y = df['Dataset']
```

```python
x.columns
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y, test_size = 0.2, random_state = 1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
algo = LogisticRegression()
algo.fit(x_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score,classification_report
confusion_matrix(y_test, algo.predict(x_test))
classification_report(y_test, algo.predict(x_test))
print(f"Accuracy is {round(accuracy_score(y_test, algo.predict(x_test))*100, 2)}%")
```
