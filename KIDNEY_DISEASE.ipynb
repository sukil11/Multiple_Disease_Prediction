```python
df['classification'] = df['classification'].replace(['ckd','notckd'], [1,0])
```

```python
df1 = df.dropna(axis = 0)
print(f"Before dropping all NaN values: {df.shape}")
print(f"After dropping all NaN values: {df1.shape}")
```

```python
df1.head()
```

```python
df1.index = range(0,len(df1),1)
df1.head()
```

```python
for i in df1['wc']:
    print(i)
```

```python
df1.info()
```

```python
df1['pcv']=df1['pcv'].astype(int)
df1['wc']=df1['wc'].astype(int)
df1['rc']=df1['rc'].astype(float)
df1.info()
```

```python
df1.info()
```

```python
object_dtypes = df1.select_dtypes(include = 'object')
object_dtypes.head()
```

```python
df1['rbc'] = df1['rbc'].map({'normal': 0, 'abnormal': 1})
df1['pc'] = df1['pc'].map({'normal': 0, 'abnormal': 1})
df1['pcc'] = df1['pcc'].map({'present': 1, 'notpresent': 0})
df1['ba'] = df1['ba'].map({'present': 1, 'notpresent': 0})
df1['htn'] = df1['htn'].map({'yes': 1, 'no': 0})
df1['dm'] = df1['dm'].map({'yes': 1, 'no': 0})
df1['cad'] = df1['cad'].map({'yes': 1, 'no': 0})
df1['appet'] = df1['appet'].map({'poor': 0,'good':1}) 
df1['pe'] = df1['pe'].map({'yes': 1, 'no': 0})
df1['ane'] = df1['ane'].map({'yes': 1, 'no': 0})
```

```python
df1.head()
```

```python
import seaborn as sns
plt.figure(figsize = (20,20))
sns.heatmap(df1.corr(), annot = True, fmt=".2f",linewidths=0.5)
```

```python
df1.corr()
```

```python
x = df1.drop(['classification', 'sg', 'appet', 'rc', 'pcv', 'hemo', 'sod'], axis = 1)
y = df1['classification']
```

```python
x.columns
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
algo = LogisticRegression()
algo.fit(x_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
confusion_matrix(y_test, algo.predict(x_test))
print(f"Accuracy is {round(accuracy_score(y_test, algo.predict(x_test))*100, 2)}%")
```

```python
import pickle
filename = 'kidney_model.sav'
pickle.dump(algo, open(filename, 'wb'))
```
